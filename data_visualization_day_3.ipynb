##### 1. Write a Python program to display a bar chart of the total bill of Gender and write the intuition about the plot.

```python
### code here
plt.figure(figsize=(5,5))
sns.barplot(data=data,x="sex",y="total_bill",hue="day")
plt.show()
```

##### 3. Write a Python program to display a violin plot of the total bill of Day and hue = "sex" and write the intuition about the plot.

```python
### code here
sns.violinplot(data=data,x="day",y="total_bill",hue="sex",split=True,showmeans=True)
plt.show()
```

on thursday females paid highest total bill below average ,males paid highest total bill above avreage 
on friday  less amount of males paid total bill,here you can see total bill is gradually increasing ,in females you can see 
normal bell curve ie total bill(data) is equally distributed

violin curve shows sharp edges on saturday with male which indicates that there are some outliers
meaning that some of males paid very high total bill from normal range ,and also very highest among all days
