```python
### code here
m=sns.FacetGrid(data=data,hue="cut",size=5)
m.map(sns.ecdfplot,"carat")
m.add_legend()
```

about diamond can diffencite between diamonds with respect to cut and carat

```python
### code here
sns.countplot(data["cut"])
```

```python
sns.scatterplot(data=data,x="carat",y="price")
```

```python
#### code here
sns.rugplot(data["depth"],height=0.3)
```

```python
sns.violinplot(data=data,x="cut",y="carat",hue="cut")
```

```python
sns.barplot(data=data,x="cut",y="carat")
```
