##### 2. Create a pie chart to show the percentage of species.

```python
### code here
plt.pie([50,50,50],labels=["Iris-setosa","Iris-versicolor","Iris-virginica"],autopct="%0.0f%%",explode=[0.1,0,0],radius=1)
plt.title("SPECIES")
plt.show()
```

##### 3. Create the joint to all numberical variable for analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
### code here
sns.jointplot(data=data,x="SepalLengthCm",y="PetalLengthCm",hue="Species")
plt.show()
```
